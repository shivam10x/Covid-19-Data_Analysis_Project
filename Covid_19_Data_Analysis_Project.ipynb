# Import Libraries___

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load DataSet___

```python
data = pd.read_csv('owid-covid-data.csv')
data.head()
```

# Data Information___

```python
data.info()
```

```python
data.describe()
```

# Data Cleaning___

```python
data.dropna()
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.info()
```

# Visualization___

# 1. How did COVID-19 cases grow over time?

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))

sns.lineplot(data = data , x = 'date' , y = 'total_cases' , palette='Set1')
plt.title('Covid-19 Cases Grow Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation = 45)
plt.show()
```

# 2. Which countries had the highest total cases?

```python

# Get latest data for each country
latest_data = data.sort_values("date").groupby("location").tail(1)

# Remove aggregate entries like 'World'
latest_data = latest_data[latest_data["iso_code"].str.len() == 3]

# Select top 10 countries by total cases
top10 = latest_data.sort_values(by="total_cases", ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(x = top10["location"],y =  top10["total_cases"] , palette='Set1')
plt.title("Top 10 Countries by Total COVID-19 Cases")
plt.xlabel("Country")
plt.ylabel("Total Cases")
plt.xticks(rotation=90)

plt.show()
```

# 3. . What is the trend in COVID-19 deaths?

```python
plt.figure(figsize=(10,6))

sns.lineplot(data = data , x = 'date' , y = 'total_deaths')
plt.title('Covid-19 Deaths Trend')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.xticks(rotation = 45)
plt.show()
```

# 4. Is there a relationship between testing and detected cases?

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data = data , x = 'total_tests' , y = 'total_cases')
```
